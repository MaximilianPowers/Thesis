```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import os
from cnn import Net
```

```python
net = Net()

eig_layer_1 = []
eig_layer_2 = []
eig_layer_3 = []
models = os.listdir('models')
for model in models:
    net.load_state_dict(torch.load('models/' + model))
    s_dict = net.state_dict()
    layer_1 = s_dict['fc1.weight'].numpy()
    layer_2 = s_dict['fc2.weight'].numpy()
    layer_3 = s_dict['fc3.weight'].numpy()

    tmp_1 = layer_1.T @ layer_1
    eig1 = np.linalg.eigvalsh(tmp_1)
    eig_layer_1.append(eig1[eig1 > 1e-5])
    tmp_2 = layer_2.T @ layer_2
    eig2 = np.linalg.eigvalsh(tmp_2)
    eig_layer_2.append(eig2[eig2 > 1e-5])
    tmp_3 = layer_3.T @ layer_3
    eig3 = np.linalg.eigvalsh(tmp_3)
    eig_layer_3.append(eig3[eig3 > 1e-5])




    
    
```

```python
fig, ax = plt.subplots(4, 3, figsize=(20, 15))
ax[0][0].hist(eig_layer_1[0], bins=50)
ax[0][0].set_title('Layer 1')
ax[0][1].hist(eig_layer_2[0], bins=40)
ax[0][1].set_title('Layer 2')
ax[0][2].hist(eig_layer_3[0], bins=5)
ax[0][2].set_title('Layer 3')

ax[1][0].hist(eig_layer_1[10], bins=50)
ax[1][0].set_title('Layer 1')
ax[1][1].hist(eig_layer_2[10], bins=40)
ax[1][1].set_title('Layer 2')
ax[1][2].hist(eig_layer_3[10], bins=5)
ax[1][2].set_title('Layer 3')

ax[2][0].hist(eig_layer_1[20], bins=50)
ax[2][0].set_title('Layer 1')
ax[2][1].hist(eig_layer_2[20], bins=40)
ax[2][1].set_title('Layer 2')
ax[2][2].hist(eig_layer_3[20], bins=5)
ax[2][2].set_title('Layer 3')

ax[3][0].hist(eig_layer_1[30], bins=50)
ax[3][0].set_title('Layer 1')
ax[3][1].hist(eig_layer_2[30], bins=40)
ax[3][1].set_title('Layer 2')
ax[3][2].hist(eig_layer_3[30], bins=5)
ax[3][2].set_title('Layer 3')
```
