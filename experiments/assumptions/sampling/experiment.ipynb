```python
import sys
sys.path.append('../../../')


import numpy as np
from riemannian_geometry.computations.sample import sample_points_heat_kernel, uniform_sample, generate_manifold_sample
from riemannian_geometry.computations.riemann_metric import LocalDiagPCA
from models.data.sklearn_datasets import MoonDataset, BlobsDataset, SpiralDataset
import matplotlib.pyplot as plt
```

```python
np.random.seed(2)

# Set the font size for matplotlib
plt.rcParams.update({'font.size': 32})
plt.tight_layout()
```

```python
dataset = MoonDataset(n_samples=1000, noise=0.05)
X = dataset.X
samples_1 = uniform_sample(10000, X)
manifold = LocalDiagPCA(X, sigma=0.1, rho=1e-5)
samples_2 = generate_manifold_sample(manifold, X, 10000, 1e-5)
samples_3 = sample_points_heat_kernel(X, 10000, t=8, connect_components=1)

fig, ax = plt.subplots(1, 3, figsize=(20*3, 16))
ax[0].scatter(samples_1[:, 0], samples_1[:, 1], c='r', s=1)
ax[0].scatter(X[:, 0], X[:, 1], c='b', s=5)
ax[0].set_title("Uniform sampling")
ax[1].scatter(samples_2[:, 0], samples_2[:, 1], c='r', s=1)
ax[1].scatter(X[:, 0], X[:, 1], c='b', s=5)
ax[1].set_title("Manifold sampling")
ax[2].scatter(samples_3[:, 0], samples_3[:, 1], c='r', s=1)
ax[2].scatter(X[:, 0], X[:, 1], c='b', s=5)
ax[2].set_title("Heat kernel sampling")
plt.savefig("figures/sampling_moon.png")
```
